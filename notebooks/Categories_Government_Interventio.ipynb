```python
#all actions
actions = ['schools_closed',
 'traveller_quarantine',
 'border_control',
 'closure_leisureandbars',
 'lockdown',
 'home_office',
 'primary_residence',
 'test_limitations']
```

```python
#country codes used in the loops below
country_codes = df_clean['adm/0'].tolist()
```

```python
start_dates_FI = []
used_actions = []
for i in actions:    
    check = (df_clean[df_clean['adm/0']=='FI'].sort_values(by='date')[i].cumsum() == 1).values[0]
    if check == False:
        continue
    else:
        start_date_FI = df_clean.iloc[(df_clean[df_clean['adm/0']=='FI'].sort_values(by='date')[i].cumsum() == 1).idxmax()].date 
        start_dates_FI.append(start_date_FI)
        if True in df_clean[df_clean['adm/0']=='FI'][i].values:
            used_actions.append(i)
#shows the start dates for actions taken in Finland; they did not do all actions
print(start_dates_FI) 
df_FL = pd.DataFrame({'FI': start_dates_FI}, index=used_actions)
df_FL
```

```python
start_dates_sc = []
cl = []
for j in country_codes:
    if j in cl:
        continue
    else:
        check = (df_clean[df_clean['adm/0']==j].sort_values(by='date')['schools_closed'].cumsum() == 1).values[0]
        if check == False:
            continue
        else:
            start_date_sc = df_clean.iloc[(df_clean[df_clean['adm/0']==j].sort_values(by='date')['schools_closed'].cumsum() == 1).idxmax()].date#[0:10]
            start_dates_sc.append(start_date_sc)
    cl.append(j)
#shows the start dates schools closed for each country
#print(start_dates_sc) 
df_sc = pd.DataFrame({'schools_closed':start_dates_sc}, index = cl)
df_sc.head()
```

```python
start_dates = []
country_list = []
for j in country_codes:
    if j in country_list:
        continue
    else:
        for i in actions:
            #see if action is done
            check = (df_clean[df_clean['adm/0']==j].sort_values(by='date')[i].cumsum() == 1).values.any()
            if check == False:
                start_dates.append(np.datetime64('NaT'))
            else:
                #getting dates actions started
                start_date = df_clean.iloc[(df_clean[df_clean['adm/0']==j].sort_values(by='date')[i].cumsum() == 1).idxmax()].date 
                start_dates.append(start_date)
                if True in df_clean[df_clean['adm/0']==j][i].values:
                    used_actions.append(i)            
               
        country_list.append(j)
#start dates for each country for each action
print(start_dates[0:8]) 
```

```python
#there is a start date for each action for each country
len(start_dates)/len(country_list) == len(actions) 
```

```python
#this is a dataframe with each country and start date for actions taken 
df = pd.DataFrame([start_dates[i:i+8] for i in range(0,len(start_dates),8)], columns = actions, index = country_list).reset_index().rename(columns={'index':'cc_index'})
df 
```

```python
country_to_codes_map = {}

for c in jhu_countries:
    if c in CC:
        country_to_codes_map[c] = df_yp_c.loc[df_yp_c['COUNTRY'] == c]["yp"].unique()[0]
    elif c in cc_jhu_map :
        c_jhu = cc_jhu_map[c]
        country_to_codes_map[c_jhu] = df_yp_c.loc[df_yp_c['COUNTRY'] == c_jhu]["yp"].unique()[0]
    
```
